```python
import numpy as np
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import StackingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import VotingRegressor
from sklearn.metrics import mean_absolute_error
```

```python
df = pd.read_csv('/content/drive/MyDrive/players_21.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.drop('club_loaned_from', axis = 1)
```

```python
df
```

```python
df = df.dropna(axis = 1)
```

```python
df
```

```python
target_variable = 'overall'
```

```python
correlation_threshold = 0.26
```

```python
correlation_matrix = df.corr()
correlation_with_target = correlation_matrix[target_variable]
```

```python
relevant_features= correlation_with_target[abs(correlation_with_target) > correlation_threshold]
```

```python
relevant_feature_names = relevant_features.index.tolist()
```

```python
data_with_relevant_features = df[relevant_feature_names]
```

```python
data_with_relevant_features
```

```python
x = data_with_relevant_features.drop('overall', axis = 1)
y = data_with_relevant_features['overall']
```

```python
kf = KFold(n_splits = 5, shuffle = True, random_state = 42)
cv_scores = []
```

```python
for train_index, test_index in kf.split(x):
    x_train, x_test = x.iloc[train_index], x.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = RandomForestRegressor()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
mse = mean_squared_error(y_test, y_pred)
```

```python
rsme = np.sqrt(mse)
```

```python
print('RSME: ', rsme)
```

```python
mae = mean_absolute_error(y_test, y_pred)
```

```python
rsme = np.sqrt(mae)
```

```python
print(mae)
```

```python
model_i = LinearRegression()
model_ii = DecisionTreeRegressor()
model_iii = RandomForestRegressor()
```

```python
voting_regressor = VotingRegressor(estimators=[('lr',model_i),('dt',model_ii),('rf',model_iii)])
```

```python
voting_regressor.fit(x_train, y_train)
```

```python
y_pred_1 = voting_regressor.predict(x_test)
```

```python
mse = mean_squared_error(y_test, y_pred_1)
```

```python
rsme = np.sqrt(mse)
```

```python
print('RSME: ', mse)
```

```python
mae = mean_absolute_error(y_test, y_pred_1)
rsme = np.sqrt(mae)
print(mae)
```
